```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
```

```python
# Load and preprocess the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, y_train = x_train[0:3000], y_train[0:3000]
x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalize pixel values to [0, 1]

y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
```

```python
from tensorflow.keras import backend as K

def my_init(shape, dtype=None):
    return K.random_normal(shape, dtype=dtype)
```

```python
# Build the neural network model
model = Sequential([
    Conv2D(2, (11, 11), activation='relu', input_shape=(28, 28, 1), use_bias=False, kernel_initializer='random_normal'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(10, activation='softmax', kernel_initializer='random_normal')     # Fully connected layer with 10 units for 10 classes and softmax activation
])

# Compile the model
sgd_optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
loss_function = tf.keras.losses.BinaryCrossentropy()
model.compile(optimizer=sgd_optimizer,
              loss=loss_function,
              metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=5, batch_size=5, shuffle=False, max_queue_size=1, validation_data=(x_test, y_test))
```
